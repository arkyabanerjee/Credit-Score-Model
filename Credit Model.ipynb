```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data_test = pd.read_csv('Test SET.csv')
data_train = pd.read_csv('Training set.csv')
```

```python
data_train.rename(columns = {'Unnamed: 0' : 'ID'}, inplace=True)
data_train.head(10)
```

```python
data_train1 = data_train.drop(['Unnamed: 0.1'], axis=1)
data_train1.head()
```

```python
data_test.rename(columns = {'Unnamed: 0' : 'ID'}, inplace=True)
data_test.head(10)
```

```python
data_test1 = data_test.drop(['Unnamed: 0.1'], axis=1)
data_test1.head()
```

```python
score =data_train1['Credit Score'].value_counts()
data = pd.DataFrame({'labels': score.index, 
                     'values': score.values})
labels = data['labels']
sizes = data['values']
colors = ['yellowgreen', 'gold', 'lightskyblue']

patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90, pctdistance=1.1, labeldistance=1.2)
plt.legend(patches, labels, loc="best")
plt.axis('equal')
plt.tight_layout()
plt.show()       
```

```python
data_train1['Credit Score'].replace({'Best':'1', 'Good':'2', 'Standard':'3'}, inplace = True)
```

```python
from sklearn.model_selection import train_test_split
y = data_train1.pop('Credit Score')
X = data_train1
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,train_size=0.8)
print("train size X : ",X_train.shape)
print("train size y : ",y_train.shape)
print("test size X : ",X_test.shape)
print("test size y : ",y_test.shape)
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.transform(X_test)
```

```python
X_test
```

```python
y_train.value_counts(normalize=True)
```

```python
from sklearn.linear_model import LogisticRegression  
classifier= LogisticRegression(random_state=42)  
classifier.fit(X_train, y_train)  
```

```python
y_pred= classifier.predict(X_test)  
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix  
cm= confusion_matrix(y_test, y_pred)  
```

```python
cm
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test, classifier.predict(X_test)))
```

```python
y_pred1 = classifier.predict(data_test1)
```

```python
y_pred1
```

```python
data_test1["Credit Score"] = y_pred1 

to_submit = data_test1[['ID',"Credit Score"]]

to_submit.to_csv("submission.csv", index = False)
```
